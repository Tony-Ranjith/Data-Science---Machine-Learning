from the above relationship graphs we can see that 'Price for sqft' is linearly correlated with rhe target variable. so we can conclude that our target variable is depend on this variable. so now we can build the model with this single feature.

This is called simple linear regression.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x=data[['Price_for_sqft']]
y=data['Sale_Price']
```

```python
x_tarin,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(x_tarin,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2=r2_score(y_test,y_pred)
r2
```

here we got higher mean squared error that means our model is not a good model. and we got that r2 square is nearly 45% that means model need to be improved.

let us plot the regression line.

```python
plt.scatter(x_test,y_test,color='blue')
plt.plot(x_test,y_pred,color='red')
plt.xlabel('Price_for_sqft')
plt.ylabel('Sale_Price')
plt.show()
```

In this notebook, we performed the exploratory data analysis (EDA) and preprocessing on the provided housing dataset. We focused on a single feature, Price_for_sqft, to understand its relationship with Sale_Price and built a basic Linear Regression model. The results, with an R-squared value of 0.45 and MSE of 8,787,257,945.59, indicate that the model is a starting point but not sufficiently powerful yet.
