```python
df=pd.read_csv('Leads.csv')
df.isnull().sum()
```

```python
leads=df.dropna()
```

```python
leads.isnull().sum()
```

```python
leads.shape
```

```python
#select numarical features
leads_num=leads.select_dtypes(include=['int64','float64'])
```

```python
#select categorical features
leads_cat=leads.select_dtypes(include=['object'])
```

```python
#will convert into numarical
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder()
leads_cat=oh.fit_transform(leads_cat)
```

```python
leads_cat.shape
```

```python
ld=pd.concat([leads_num,pd.DataFrame(leads_cat.toarray())],axis=1)
```

```python
ld.head()
```

```python
ld.isnull().sum()
```

```python
ld=ld.dropna()
```

```python
ld
```

```python
ld.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
x=ld.drop('Converted',axis=1)
y=ld['Converted']
```

```python
x=x.astype('int')
```

```python
y=y.astype('int')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
std=StandardScaler()
```

```python
x_train.columns = x_train.columns.astype(str)
x_test.columns = x_test.columns.astype(str)
x_train=std.fit_transform(x_train)
x_test=std.transform(x_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#roc curve
from sklearn.metrics import roc_curve
```

```python
roc_curve(y_test,y_pred)
```

```python
#classification report
from sklearn.metrics import classification_report
classification_report(y_test,y_pred)
```

```python
#will plot now model sigmoid curve
fpr,tpr,threshold=roc_curve(y_test,y_pred)
plt.plot(fpr,tpr)
```
