```python
df=load_breast_cancer(as_frame=True)
```

```python
data=df['data'] #convert into dataframe
data['target']=df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.svm import SVC
```

```python
x=data.drop('target',axis=1)
```

```python
y=data['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
model=SVC(kernel='linear',C=1.0,gamma='auto')
```

```python
model.fit(x_train,y_train)
print('model training cpmpleted')
```

```python
y_pred=model.predict(x_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

our model classify the 41 of class 0 are true and 2 of it miss classified

and 68 truely classified and 3 misclassified of class 1

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
print("accuracy score",accuracy_score(y_test,y_pred))
```

Our model is 95% accurate. that means it can classify 95% truely
